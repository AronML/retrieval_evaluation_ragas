```python
import pandas as pd
```

```python
results_hf = pd.read_parquet("results/results_HF_embeddings_split.parquet")
results_google = pd.read_parquet("results/results_Google_embeddings_split.parquet")
results_openai = pd.read_parquet("results/results_OpenAI_embeddings_split.parquet")
```

```python
results_openai['provider'] = 'OpenAI'
results_google['provider'] = 'Google'
results_hf['provider'] = 'HuggingFace'
```

```python
results = pd.concat([results_google, results_openai, results_hf],axis=0)
```

```python
results
```

```python
results.groupby(['provider']).agg({
    'context_precision': 'mean',
    'context_recall': 'mean',
    'context_entity_recall': 'mean'}).reset_index()
```
